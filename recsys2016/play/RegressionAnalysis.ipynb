```python
import pandas as pd
import os
```

```python
appName = 'beautySpa'
# appName = 'hotel'
# appName = 'restaurant'
 
path = '../../../data/'
```

```python
# 1) load data
reviews_file = os.path.join(path,'BING_DATASET', appName+'_data.csv')
df = pd.DataFrame.from_csv(reviews_file)
```

```python
df.columns
```

```python
for c in df.columns:
    print(c,'\t',df[c].unique(),'\t',df[c].value_counts())
```

```python
asp = 'EXPERIENCE_WAIT'
overall = asp+'sent'
df[overall] = df[overall].fillna(-10.0)

print(overall,'\t',df[overall].unique(),'\n',df[overall].value_counts())
df.groupby(overall)['stars'].mean()
```
